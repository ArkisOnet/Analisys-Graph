```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import date, datetime, time
```

```python
spotify_list=pd.read_csv('data.csv')
spotify_list.head()
```

```python
avr=spotify_list.groupby('year')['popularity'].mean().reset_index()
```

```python
plt.figure(figsize=(10,6))
plt.plot(avr['year'],avr['popularity'],marker='o',label='Avg popularity')
plt.title('Popularity of tracks by year')
plt.xlabel('Year')
plt.ylabel('Popularity')
plt.grid(True)
plt.legend()
plt.show()
```
